```python
from functools import reduce
from collections import Counter

def hashtags(tweet):
    return list(map(lambda x: x.lower(), filter(lambda x: x[0] == "#", str(tweet).split())))
def flatten(data):
    return Counter(reduce(lambda x,y: x + y, list(map(lambda x: hashtags(x), data.tolist()))))
```
